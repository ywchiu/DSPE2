```python
import re
print re.match(r'[9]' , '9')
print re.match(r'[7-9]{3}' , '987')
print re.match(r'[0-9]+' , '192048274987')
print re.match(r'[0-9]+' , '') # not match
print re.match(r'[0-9]*' , '')
print re.match(r'\d*' , '') #[0-9] = \d
print re.match(r'[a-z]+' , 'hello')
print re.match(r'[a-z]+' , 'Hello') # not match
print re.match(r'[a-zA-Z]+' , 'Hello')
print re.match(r'\w+' , 'Hello')
print re.match(r'\w+' , 'Hela123dsfdf')
print re.match(r'\w+' , 'Hela123[]d!  @#sfdf')
print re.match(r'.+' , 'Hela123[]d!  @#sfdf')



```

```python
# -*- coding: utf-8 -*- 
from bs4 import BeautifulSoup 
import re, requests 
import time
#import os
#os.mkdir('gov')
bid_list = open('bid_list.txt', 'r') 
for line in bid_list.readlines(): 
    pagelink = line.strip() 
    m = re.match(r"([^ ]+)pkAtmMain=(?P<pkAtmMain>.+)&tenderCaseNo=(?P<tenderCaseNo>.+)", pagelink) 
    filename = "%s_%s"%(m.group('pkAtmMain'), m.group('tenderCaseNo')) 
    request_get = requests.get(pagelink) 
    response = request_get.text.encode('utf8') 
    soup = BeautifulSoup(response) 
    printarea = soup.select('#printArea')[0] 
    #print filename
    bid_detail = open("gov/%s.txt"%(filename), 'w') 
    bid_detail.write(printarea.prettify("utf-8")) 
    bid_detail.close() 
    time.sleep(0.01)
bid_list.close()
```
